```python
# Core Libraries
import pandas as pd
import numpy as np
# Visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load dataset
df = pd.read_csv('HRDATA.csv')
df.head()
```

```python
#dataset shape
print("Dataset shape:", df.shape)
```

```python
#dataset info
df.info()
```

```python
#Descriptive Statistics
df.describe()
```

```python
# Separate variable types
num_cols = df.select_dtypes(include=["int64", "float64"]).columns
cat_cols = df.select_dtypes(include=["object"]).columns

print("Numerical Variables:", len(num_cols))
print("Categorical Variables:", len(cat_cols))
```

Will need to encode the cat varibles later when building model

```python
# Check for constant columns
constant_cols = [col for col in df.columns if df[col].nunique() == 1]
constant_cols
```

Checked for columns with no variance, this means that these columns will not have any prediction power

```python
# Target distribution (counts and proportions)
attrition_counts = df["Attrition"].value_counts()
attrition_ratio = df["Attrition"].value_counts(normalize=True)

print("Counts:")
print(attrition_counts)
print("\nProportions:")
print(attrition_ratio)
# Visualize target distribution
plt.figure(figsize=(10, 6))
sns.countplot(x="Attrition", data=df, hue="Attrition", palette="viridis", legend=False)
plt.title("Attrition Distribution")
plt.xlabel("Attrition")
plt.ylabel("Count")
```

Outcome variable is imbalanced, will need to look at precision and recall metrics instead of accruacy when comparing models. 

```python
# Group by Attrition and calculate summary stats for MonthlyIncome
income_stats = df.groupby('Attrition')['MonthlyIncome'].describe()
print(income_stats)

plt.figure(figsize=(10, 6))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df, palette='viridis', hue='Attrition', dodge=False, legend=False)
plt.title('MonthlyIncome Distribution by Attrition Status')
plt.xlabel('Attrition')
plt.ylabel('MonthlyIncome')
plt.show()
```

Significant difference of income (5204 vs 3202) between attrition outcomes. Inference: If you are compensated better you are less likley to leave

```python
# Mean comparison by Attrition
df.groupby("Attrition")[num_cols].mean().T
```

```python
#Attrition by JobLevel
print(df.groupby("Attrition")["JobLevel"].mean())
# Calculate and print percentage of attrition for each job level
joblevel_counts = df.groupby(['JobLevel', 'Attrition']).size().unstack(fill_value=0)
attrition_percentage = (joblevel_counts['Yes'] / joblevel_counts.sum(axis=1)) * 100
print("Percentage of Attrition for each JobLevel:")
print(attrition_percentage)

plt.figure(figsize=(10, 6))
sns.countplot(x="JobLevel", hue="Attrition", data=df, palette="viridis")
plt.title("Attrition by Job Level")
plt.xlabel("JobLevel")
plt.ylabel("Count")
plt.show()
```

Strong Negative cor. Inferance: Entry level more likley to leave vs Higher level. Backs MonthlyIncome hypo. (Corr vs Causal): Entry level paid less vs Highler Level Paid more. Employees in early early stages of their careers are more likley to look for career advancement. Level 1 and 3 had highest attrition rate with 26% & 14%, Level 2 had high rention rate (91%), 4 and 5 Low attrition rate.

```python
#Attrition by StockOptionLevel
print(df.groupby("Attrition")["StockOptionLevel"].mean())
# Calculate and print percentage of attrition for each stock option level
stockoption_counts = df.groupby(['StockOptionLevel', 'Attrition']).size().unstack(fill_value=0)
stockoption_attrition_percentage = (stockoption_counts['Yes'] / stockoption_counts.sum(axis=1)) * 100
print("Percentage of Attrition for each StockOptionLevel:")
print(stockoption_attrition_percentage)

plt.figure(figsize=(10, 6))
sns.countplot(x="StockOptionLevel", hue="Attrition", data=df, palette="viridis")
plt.title("Attrition by Stock Option Level")
plt.xlabel("StockOptionLevel")
plt.ylabel("Count")
plt.show()
```

Strong Negative Cor. between stockoption level and attrition. Level 0 high attrition rate (24%), Level 1 and 2 significantly lower, but level 3 back up (17%). Level 3 is not very significant since there is so little employees. Level 0 vs rest has huge predicting potential. Possible grouping later in feature engineering.
